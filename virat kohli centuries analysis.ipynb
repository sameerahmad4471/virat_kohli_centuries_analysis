```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("71 Centuries of Virat Kohli.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Unnamed: 14"],inplace=True,axis=1)
```

```python
df.info()
```

```python
df.columns
```

# Centuries in All Formats

```python
ax =sns.countplot(x="Format",data=df)
sns.set(rc={"figure.figsize":(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# odi and test centuries
```

# Centuries Against Teams

```python
ax =sns.countplot(y="Against",data=df)
sns.set(rc={"figure.figsize":(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#more centuries against Australia Srilanka and West Indies
```

# centuries against teams in Home and Away

```python
ax =sns.countplot(x="Against",data=df,hue="H/A")
sns.set(rc={"figure.figsize":(15,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# Won or Loss

```python
ax =sns.countplot(x="Result",data=df,hue="Format")
sns.set(rc={"figure.figsize":(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#most centuries in won cause
```

# Centuries as a Captain

```python
ax =sns.countplot(x="Captain",data=df)
sns.set(rc={"figure.figsize":(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#more centuries as a captain
```

# Home or Away

```python
ax =sns.countplot(x="H/A",data=df,hue="Format")
sns.set(rc={"figure.figsize":(10,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# centuries in inning_no

```python
ax =sns.countplot(x="Inn.",data=df)
sns.set(rc={"figure.figsize":(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#most centuries in 2 inning
```

# Man of the Match

```python
ax =sns.countplot(x="Man of the Match",data=df)
sns.set(rc={"figure.figsize":(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
